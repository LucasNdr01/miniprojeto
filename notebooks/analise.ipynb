```python
#Importando bibliotecas
import pandas as pd
import numpy as np
```

```python
#Leitura do arquivo CSV
df = pd.read_csv('../dados/base_varejo_limpa.csv', sep=';')
```

```python
#Analisando categorias vendidas por gênero 
agrupamento_genero = df.groupby(['PR_CAT', 'CL_GENERO']).agg(
    total_vendas=('CO_ID', 'count')
)

agrupamento_genero
```

```python
#Analisando produtos vendidos por gênero
produto_genero = (
    df.groupby(['PR_NOME', 'CL_GENERO'])
    .agg(vendas=('CO_ID', 'count'))
    .reset_index()
    .sort_values(by='vendas', ascending=False)
)

produto_genero.head(15)
```

```python
#Estatísticas descritivas da coluna de número de filhos do cliente

estatisticas = pd.DataFrame({
    'Métrica': [
        'Média',
        'Mediana',
        'Desvio padrão',
        'Moda',
        'Mínimo',
        'Quartil 1 (25%)',
        'Quartil 2 (50%)',
        'Quartil 3 (75%)',
        'Máximo',
        'Contagem'
    ],
    'Valor': [
        df['CL_FHL'].mean(),
        df['CL_FHL'].median(),
        df['CL_FHL'].std(),
        df['CL_FHL'].mode().iloc[0],
        df['CL_FHL'].min(),
        df['CL_FHL'].quantile(0.25),
        df['CL_FHL'].quantile(0.50),
        df['CL_FHL'].quantile(0.75),
        df['CL_FHL'].max(),
        df['CL_FHL'].count()
    ]
})

estatisticas
```
